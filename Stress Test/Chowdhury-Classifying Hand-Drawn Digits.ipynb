# Classifying Hand-Drawn Digits

Since we are making our OC algorithm application specific to training neural networks, we need to write a stress program to stress test our GPU for accuracy and stability after successful application of each OC profile generated by our algorithm. Our stress program stresses the GPU for accuracy and stability by training a fully connected feed forward neural network for the classification of hand drawn digits using the cross entropy loss function for 20 iterations. The training dataset is an MNIST dataset of hand drawn digits.

In this section we briefly go over the methodologies used to obtain the network architecture used for stress testing.

```python
# Adapted from: 
#  https://stackoverflow.com/questions/287871/print-in-terminal-with-colors

class bcolors:
    LINE = '-------------------------------------------------'
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

import numpy as np
import gzip, sys, pickle

# Do the necessary imports
import numpy as np
from numpy import sin as sin
from numpy import cos as cos
from numpy import tan as tan
from numpy import exp as e
from numpy import sqrt as sqrt
from numpy import pi as pi
from numpy import log
from copy import deepcopy
import matplotlib.pyplot as plt
from IPython.display import display, Math
from prettytable import PrettyTable
from IPython.display import HTML
from scipy.special import erfi as erfi
from scipy.special import erf as erf
import IPython.display as ipd
import pandas as pd
import time, copy, torch, subprocess

with gzip.open('mnist.pkl.gz', 'rb') as f:
    train_set, valid_set, test_set = pickle.load(f, encoding='latin1')
    
Xtrain = train_set[0].reshape((-1, 1, 28, 28))
Ttrain = train_set[1]
Xtest = test_set[0].reshape((-1, 1, 28, 28))
Ttest = test_set[1]

Xtrain.shape, Ttrain.shape, Xtest.shape, Ttest.shape

import neuralnetworks_pytorch as nn

def percent_correct(actual, predicted):
    return 100 * np.mean(actual == predicted)
```

## Finding the network architecture for stress testing

```python
best_value_n_iter = 800 #int(tmp_res[np.argmax(tmp_res[:, 1]), 0])
print("best number of iterations from part 1:", best_value_n_iter)

print("Trying out the following hidden layer structures")
print("Each row in the following text is a hidden layer")
print("structure.\n")

np.random.seed(370)
n_hiddens = [np.random.randint(1, 1000, np.random.randint(5, 15)) for i in range(16)]

for nh in n_hiddens:
    print(list(nh))
```

```python
batch_size = 50000
learning_rate = 0.001
n_hiddens = n_hiddens
n_iters = best_value_n_iter

res = pd.DataFrame(columns=['Hidden Layer Struct.', 'Percentage Correct'])

start_time = time.time()

training_accuracy = []
training_time = []
memory_usage = []
gpu_core_utilization_over_time = []
gpu_memory_utilization_over_time = []
cmd = 'nvidia-smi --query-gpu=memory.used --format=csv,noheader,nounits'

for i,nh in enumerate(n_hiddens):

    print(bcolors.OKGREEN)
    print(bcolors.LINE)
    print("Trying out hidden layer struct.:")
    print(nh)
    print(bcolors.LINE)
    print(bcolors.ENDC)

    nnet_classify = nn.NeuralNetworkClassifier_Pytorch(np.prod(Xtrain.shape[2:]),
                                                       nh, 
                                                       10,
                                                       relu=False, 
                                                       gpu=True,
                                                       n_conv_layers=0,
                                                       windows=[], 
                                                       strides=[],
                                                       input_height_width=None)
    
    _initial_weights_ = copy.deepcopy(nnet_classify.dumpWeights())
    
    trainAcc, gpuDataFile = nnet_classify.train(Xtrain, 
                                                Ttrain, 
                                                Xtest, 
                                                Ttest, 
                                                n_iters, 
                                                batch_size, 
                                                learning_rate)
    
    training_accuracy += [trainAcc]
    memory_usage += [
        int(subprocess.check_output(cmd, shell=True).decode().split('\n')[0])
    ]
    training_time += [
        nnet_classify.time
    ]
    gpuData = np.load("/tmp/" + gpuDataFile + ".npy")
    gpu_core_utilization_over_time += [gpuData[0]]
    gpu_memory_utilization_over_time += [gpuData[1]]
    
    res.loc[len(res)] = [nh, 
                            percent_correct(Ttest, 
                            nnet_classify.use(Xtest)[0])]

    print(bcolors.OKBLUE)
    print(bcolors.LINE)
    print("Training for", n_iters, 'iterations')
    print("Took:", nnet_classify.time, 'seconds')
    print("Memory Usage:", memory_usage[-1], "MB")
    print(bcolors.LINE)
    print('Progress:', np.round((i+1)/len(n_hiddens)*100, 5), "%")
    print(bcolors.ENDC)


print("Total time taken:", time.time()-start_time, 'seconds')

_final_weights_ = nnet_classify.dumpWeights()



#-------------------------------------------------------------------------
(
    torch.sum(_initial_weights_[0].weight.data),
    torch.sum(_final_weights_[0].weight.data)
)
```

### GPU and Memory utilization for each structure

```python
nrows, ncols = 1, 2
for i in range(len(gpu_core_utilization_over_time)):
    if i%ncols == 0:
        plt.figure(figsize=(17, 5))
    
    plt.subplot(nrows, ncols, (i%ncols)+1)
    plt.title("Utilization Vs. Time for \n" + str(n_hiddens[i]), fontsize=14)
    plt.xlabel('Time (seconds)', fontsize=11)
    plt.ylabel('Utilization (%)', fontsize=11)
    plt.plot(gpu_core_utilization_over_time[i], '-or', label='GPU Utilization', linewidth=1)
    plt.plot(gpu_memory_utilization_over_time[i], '-og', label='Memory Utilization', linewidth=1)
    plt.legend()
    
    if i%ncols == nrows:        
        plt.show()
        print()
```

From the above plot, it can be seen that all of randomly generated neural network structures have a GPU utilization value of close to 100% while the memory utilization value is in the range of 50 to 60%.

### Accuracy of each structure

```python
y = res.values[:, 1]

plt.figure(figsize=(15,8))
plt.ylabel('Percentage Correct', fontsize=12)
plt.title('Hidden Layers Structure Vs. Percentage Correct', fontsize=14)
plt.plot(y, '--o', linewidth=1.0, markersize=9, label='Structure')
plt.xticks(range(y.shape[0]), n_hiddens, rotation=25, horizontalalignment='right')
plt.xlabel('Hidden Layers Structure', fontsize=14)
plt.legend()
plt.show()
```
